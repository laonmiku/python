```python
import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
def add_cm(height):
  return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def lower(lang):
  if pd.notnull(lang):
    return lang.lower()
  else:
    return lang
  
df['SW특기'] = df['SW특기'].apply(lower)
df
```
