```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시 마이너스 글자가 깨지는 현상을 해결

import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
x = list(df['영어'])
y = list(df['수학'])
sizes = list(df['학년'] * 500)
sizes
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
for idx, no in enumerate(df['이름']):
  plt.text(x[idx], y[idx]+1, no, size=8, ha='center', color='green')
plt.colorbar(ticks=[1,2,3], orientation='horizontal', label='학년', shrink=0.5)
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(15, 10))
fig.suptitle('여러그래프')
a1=axs[0,0]
a1.bar(df['이름'], df['키'])
a1.set_title('첫번째 그래프')
a1.set(xlabel='이름', ylabel='키', ylim=(150, 210))
a1.grid(ls='--')
a1.set_facecolor('lightyellow')

a2 = axs[0,1]
a2.plot(df['이름'], df['수학'], label='수학')
a2.plot(df['이름'], df['영어'], label='영어')
a2.plot(df['이름'], df['사회'], label='사회')
a2.legend()

a3 = axs[1,0]
a3.barh(df['이름'], df['키'])

a4 = axs[1,1]
a4.scatter(df['영어'], df['수학'])
```
